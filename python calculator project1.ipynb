```python
# ==============================
# REAL CALCULATOR APP FOR GOOGLE COLAB
# Beautiful UI + GUI + UX
# ==============================

import ipywidgets as widgets
from IPython.display import display, HTML

# ------------------------------
# Custom CSS Styling
# ------------------------------

display(HTML("""
<style>
.calc-box {
    background: #1e1e2f;
    padding: 25px;
    border-radius: 20px;
    width: 320px;
    margin: auto;
    box-shadow: 0px 0px 20px rgba(0,0,0,0.4);
}

.calc-title {
    color: white;
    text-align: center;
    font-size: 28px;
    font-weight: bold;
    margin-bottom: 15px;
}

.widget-button {
    height: 55px !important;
    width: 65px !important;
    font-size: 20px !important;
    border-radius: 12px !important;
}

.output-box {
    background: black;
    color: lime;
    font-size: 28px;
    text-align: right;
    padding: 10px;
    border-radius: 10px;
    margin-bottom: 15px;
}
</style>
"""))

# ------------------------------
# Display Area
# ------------------------------

display_box = widgets.Text(
    value='',
    placeholder='0',
    disabled=True,
    layout=widgets.Layout(width='290px', height='55px')
)

display_box.add_class("output-box")

# ------------------------------
# Calculator Logic
# ------------------------------

expression = ""

def on_button_click(b):
    global expression

    value = b.description

    if value == "C":
        expression = ""

    elif value == "=":
        try:
            expression = str(eval(expression))
        except:
            expression = "Error"

    else:
        expression += value

    display_box.value = expression

# ------------------------------
# Buttons
# ------------------------------

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+',
]

button_widgets = []

for text in buttons:
    button = widgets.Button(
        description=text,
        layout=widgets.Layout(width='65px', height='55px')
    )

    button.on_click(on_button_click)
    button_widgets.append(button)

# Clear Button
clear_btn = widgets.Button(
    description='C',
    layout=widgets.Layout(width='290px', height='55px')
)

clear_btn.on_click(on_button_click)

# ------------------------------
# Layout Design
# ------------------------------

grid = widgets.GridBox(
    children=button_widgets,
    layout=widgets.Layout(
        grid_template_columns="repeat(4, 70px)",
        grid_gap="10px"
    )
)

calculator_ui = widgets.VBox([
    widgets.HTML("<div class='calc-title'>Calculator App</div>"),
    display_box,
    grid,
    clear_btn
])

# ------------------------------
# Final Display
# ------------------------------

display(calculator_ui)
```
